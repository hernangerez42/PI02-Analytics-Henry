```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Accidentes_Aviones_mod.csv')
```

```python
df.describe()
```

```python
#Realizamos una grafico de dispersion entre "todos_abordo" y "cantidad_fallecidos"
sns.scatterplot(x="todos_abordo", y="cantidad_fallecidos", data=df)
```

```python
#Grfico de dispersin entre "todos_abordo" y "decada"
sns.scatterplot(x="todos_abordo", y="decada", data=df)
```

```python
# Graficar un histograma de la columna "cantidad_fallecidos"
sns.distplot(df['cantidad_fallecidos'], kde=False)
```

```python
# Crear un histograma de todas las columnas en el DataFrame 'df'
df.hist(bins = 50, figsize = (20,15))
plt.show
```

```python
# Crear un grfico de dispersin entre las columnas "todos_abordo" y "cantidad_fallecidos" en el DataFrame 'df'
df.plot(kind = "scatter", x = "todos_abordo", y = "cantidad_fallecidos", alpha = 0.1)
```

```python
# Crear un grfico de dispersin entre las columnas "todos_abordo" y "sobrevivientes" en el DataFrame 'df'
df.plot(kind = "scatter", x = "todos_abordo", y = "sobrevivientes", alpha = 0.1)
```

```python
columnas = ['todos_abordo', 'pasajeros_abordo', 'tripulacion_abordo', 'cantidad_fallecidos', 'pasajeros_fallecidos', 'tripulacion_fallecidos', 'fallecidos_en_tierra', 'sobrevivientes']
submatriz = df[columnas]
matriz_correlacion = submatriz.corr()
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm')
plt.figure(figsize=(10, 8))
plt.show()
```
